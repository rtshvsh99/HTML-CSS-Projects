```python
#classMethod ----> Whenever we have deal with class variable then we can use the class method 
# Calling class method----->  we call the class method by using class name , we can also call it by using object then whats the need of class mathod so we call by className

#class method without parameter

class Mobile:
  fp = 'Yes'                 #-->   class variable / static variable
  @classmethod        #Decorator
  def show_model(cls):        #Class method
    print("RealMe X","Fingerprint:",cls.fp)

realme = Mobile()
Mobile.show_model()         #Calling Class method

```

```python
#class method with parameter

class Mobile:
  fp = 'Yes'                 #-->   class variable / static variable
  @classmethod        #Decorator
  def show_model(cls,r):        #Class method
    cls.ram = r
    print("RealMe X","Fingerprint:",cls.fp)
    print("Ram:",cls.ram)

realme = Mobile()
Mobile.show_model('4GB')         #Calling Class method
```

```python
#Static Method with and without parameter
class Mobile:
 
  @staticmethod
  def show_model(p,m):
    price = p
    model = m
    print("Model:",model, "Price:",price)

realme = Mobile()
Mobile.show_model(10000,"RealMe 3")
```

```python
#------------------------------>>>>>>>>>>-Perfect example of Static method----------------------<<<<<<<<<<<<<<<----------------->

class Student:
  def __init__(self,n,r):
    self.name = n
    self.roll = r
  # instance method
  def disp(self):
    print("Student Name:",self.name)
    print("Student Roll No.-:",self.roll)

class User:
  #static method
  @staticmethod
  def show(s):
    print("User Name:",s.name)
    print("User roll:",s.roll)
    s.disp()

stu = Student("Ramesh",102)
User.show(stu)
```

```python
#Inheritance 

#Single Inheritance

class Father():
  money = 1000   #class Variable

  def show(self):
    print("Parent Class Instance Method")

  @classmethod
  def showmoney(cls):
    print("Parent Class class method",cls.money)

  @staticmethod
  def stat():
    a = 10
    print("Parent class Static method",a)

class Son(Father):
  def disp(self):
    print("Child class Instance method")

s = Son()
s.disp()
s.show()
s.showmoney()
s.stat()



```

```python
#Construcor with super method without parameter
class Father:
  def __init__(self):
    self.money=1000
    print("Parent class Constructor")

  def show(self):
    print("Parent class Instance method")

class Son(Father):
  def __init__(self):
    super().__init__()  #Calling parent class Constructor
    
    print("Child Class Construcor")

  def disp(self):
    print("Child class Instance method",self.money)

s = Son()
s.disp()
```

```python
#Construcor with super method with parameter
class Father:
  def __init__(self,m):
    self.money=m
    print("Parent class Constructor")

  def show(self):
    print("Parent class Instance method")

class Son(Father):
  # def __init__(self,m):
  def __init__(self,m,j):
    super().__init__(m)      #Calling parent class Constructor
    self.job = j
    print("Child Class Construcor")

  def disp(self):
    print("Child class Instance method",self.money,"Job:",self.job)

s = Son(1000,"Engineer")
s.disp()
```

```python
#Multilevel Inheritance
class Father:
  def __init__(self):
    print("Father Class Constructor")
  def showF(self):
    print("Father Class Instance Method")

class Son(Father):
  def __init__(self):
    super().__init__()   #Calling Father class Constructor
    print("Son Class Constructor")
  def  showS(self):
    print("Son Class Instance Method")

class GrandSon(Son):
  def __init__(self):
    super().__init__()    #Calling Son class Constructor
    print("GrandSon Class Constructor")
  def showG(self):
    print("GrandSon class Instance Method")

g = GrandSon()
# g.showF()
# g.showS()
# g.showG()
```

```python
#Strong Typing

class Duck:
  def walk(self):
    print("thapak,thapak,thapak,thapak")
class Horse:
  def walk(self):
    print("tapak,tapak,tapak,tapak")


class Cat:
  def talk(self):
    print("Mou,Mou,Mou,Mou")


def myfunction(obj):
  if hasattr(obj,'walk'):
    obj.walk()
  if hasattr(obj,'talk'):
    obj.talk()

d = Duck()
myfunction(d)
h = Horse()
myfunction(h)

c = Cat()
myfunction(c)
```
